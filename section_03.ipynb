# Create 2D charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

# Line Charts

```python
np.random.seed(40)
x = np.linspace(10,24,10)
y = np.random.randn(10)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
lines = ax.plot(x, y)

# for more features help(ax.plot)
```

```python
fig, ax = plt.subplots(figsize = (12,4))
ax.plot('Sepal length', data =df, label = 'Sepal')
ax.plot('Petal length', data =df, label = 'Petal')
ax.legend()
```

```python
ax.set_title('Iris', size = 22)
ax.set_xlabel('Index', size =14)
ax.set_ylabel('Length', size = 14)
ax.tick_params(axis = 'both',labelsize = 12, labelcolor = 'green')
from matplotlib.ticker import StrMethodFormatter
Str_f = StrMethodFormatter("{x:.1f} inch")
ax.yaxis.set_major_formatter(Str_f)
fig
```

```python
line1 = ax.lines[0]
line1.set(marker = 'o', markevery =2, markerfacecolor = 'black')
fig
```

```python
value = df['Petal length'].values
pmax = value.max()
pmin = value.min()
xmin = value.argmin()
xmax = value.argmax()

ax.axvline(xmin, color ='red', ls = 'dashed')
ax.axvline(xmax, color ='red', ls = 'dashed')
fig
```

```python
min_anot = ax.annotate(str(pmin), xy =(xmin,pmin), xytext =(xmin+10, pmin+2), arrowprops = dict(arrowstyle = '->') )
min_anot.set(size = 22, color = 'white')
min_anot.set_bbox(dict(facecolor = 'red', boxstyle = 'round, pad = 0.4', lw = 0))
min_anot.arrow_patch.set_connectionstyle('angle, angleA=-90')
min_anot.arrow_patch.set(lw = 2, linestyle = 'dashed', color = 'red')
fig
```

```python
fig, ax1 = plt.subplots()

df[['Sepal length','Petal length']].plot.line(ax = ax1)
```

# Bar charts

```python
df = pd.read_csv('barcharts-us-music-sales-by-genre.csv',index_col = 'Year')
df.head()
```

### Without Pandas

```python
fig, ax = plt.subplots(figsize = (12,4))

rock = ax.bar(df.index, 'Rock', data=df, label = 'Rock %')
jaaz = ax.bar(df.index, 'Jazz', data=df, label = 'Jazz %')
ax.legend()
```

```python
ax.set_title('Rock vs Jazz',size = 22)
ax.set_xticks(df.index)
fig
```

```python
spines = ['bottom','top','left','right']
for s in spines:
  ax.spines[s].set_visible(False)
ax.legend(frameon = False)
fig
```

```python
ax.tick_params(left = False,bottom = False)
fig
```

```python
for p in rock.patches:
  t = ax.annotate(str(p.get_height())+'%',(p.get_x(),p.get_height()+2))
  t.set(color='black',size =8)
  t.set_bbox(dict(facecolor = 'white', boxstyle = 'circle', lw = 1, edgecolor = 'red'))
fig
```

# Bar charts-Grouped

```python
width = 0.3

fig,ax = plt.subplots(figsize = (12,4))
rock = ax.bar(df.index, 'Rock',data = df, label = 'Rock %', width = width)
jazz = ax.bar(df.index+width, 'Jazz', data=df, label = 'Jazz %', width =width)
```

```python
ax.set_xticks(df.index)
fig
```

```python
ax.set_xticks(df.index+width/2)
ax.set_xticklabels(df.index)
fig
```

### In simple way using pandas

```python
fig2, ax2 = plt.subplots(figsize = (12,4))
df[['Rock','Jazz']].plot.bar(ax = ax2)
ax2.set_title('Rock vs Jazz',size =22)
ax2.tick_params(labelrotation = 0)
```

```python
for p in ax2.patches:
  t = ax2.annotate(str(p.get_height())+'%',(p.get_x(),p.get_height()+1))
  t.set(color='black',size =5)
  t.set_size(10)
fig2
```

```python
ax1, ax2 = df[['Rock','Jazz']].plot.bar(subplots = True,figsize = (12,4))
ax2.tick_params(labelrotation = 0)
```

# Pie charts

```python
df = pd.read_csv('barchart-divorcerates.csv')
```

```python
df.head()
```

```python
fig,ax = plt.subplots()
df.plot.pie(y='Women',labels = df.Age, autopct = '%1.1f%%',
            wedgeprops = dict(ls = 'dotted', ec = 'white', lw = 2),
            textprops = dict(color = 'white',fontsize = 12),ax = ax)
ax.set_aspect('equal')
ax.set_ylabel('')
ax.legend_.set_visible(False)
ax.set_title('Divorce women rate by age group',size = 20,color ='white')
fig.set_facecolor('#150c54')
```

# Polar Charts

```python
df = pd.read_csv('barchart-divorcerates.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize = (12,6))

x = df.Age
y1 = df.Women
y2 = df.Men

ax.bar(x, (y1+y2), label = 'Women', alpha = 0.8)
ax.bar(x, y2, label = 'Men', alpha = 0.8)

ax.tick_params(labelleft = 'Off', left = 'Off')
l = ax.legend(loc = 'upper right')
l.set_bbox_to_anchor((1, 1))
```

```python
fig, ax = plt.subplots(figsize = (12,6),subplot_kw = dict(polar = True))

x = df.Age
y1 = df.Women
y2 = df.Men

ax.bar(x, (y1+y2), label = 'Women', alpha = 0.8)
ax.bar(x, y2, label = 'Men', alpha = 0.8)

ax.tick_params(labelleft = 'Off', left = 'Off')
l = ax.legend(loc = 'upper right')
l.set_bbox_to_anchor((1.2, 1.1))
```

```python
ax.set_thetamin(-45)
ax.set_thetamax(90)
fig
```

```python
import numpy as np

p = np.pi
ax.set_xbound(0, p)
fig
```

# Bar and then Dual axis

```python
df = pd.read_csv('sales.csv')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize = (12,6))

ax.bar(df.index, df.Quantity, label = 'Quantity')
ax.set_xticks(df.index)
ax.set_xticklabels(df.Date)
ax.set_facecolor('#18104f')
fig.set_facecolor('#34297d')
```

```python
ax.set_title('Sales by month',size = 20, color = 'white')
ax.tick_params(colors = 'white', bottom = 'Off', left = 'Off', axis = 'y', labelsize =14)
for p in ax.patches:
  t = ax.annotate(str(p.get_height()),(p.get_x(),p.get_height()+10))
  t.set(color='white',size = 5)
  t.set_size(10)
fig
```

# Dual axis charts
