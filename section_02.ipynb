#Add Line2D Artist

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
np.random.seed(40)
```

```python
ash = np.linspace(1, 40, 10)
y1 = ash/(np.random.rand(10)*2)
y2 = ash/(np.random.rand(10)*5)
```

```python
line1 = Line2D(ash, y1)
line2 = Line2D(ash, y2)

fig, ax = plt.subplots()

ax.add_line(line1)
ax.add_line(line2)
```

```python
ax.autoscale()
fig
```

```python
ax.lines[0].set_color('y')
fig
```

```python
fig.set_size_inches(15,7)
fig
```

```python
l1 = ax.lines[0]
l2 = ax.lines[1]

# change l1 opacity by
l1.set_alpha(0.5)
fig
```

```python
for line in ax.lines:
  line.set_color('r')
  line.set_lw(2)
fig
```

```python
l2.set_marker("^")
l2.set_markersize(25)
l2.set_markerfacecolor("white")
fig
```

```python
l1.set_linestyle("dashed")
fig
```

```python
l2.set_sketch_params(0.8)
fig
```

```python
l1.set_clip_on(False)
ax.set_xlim(0,40)
fig
```

```python
ax.set_title("This a line plot project task")
ax.set_xlabel("X axis title")
ax.set_ylabel("Y axis title")
fig
```

## Rectangle Properties

```python
np.random.seed(20)

x = list(range(1,6))
y = np.linspace(1,10,5)
```

```python
fig, ax = plt.subplots()

def BarBanai(ax, x, y, w):
  for x, y in zip(x, y):
    bar =Rectangle((x, 0), w, y)
    ax.add_patch(bar)

BarBanai(ax, x, y, 0.4)
ax.autoscale()
```

```python
fig.set_size_inches(15,7)
ax.set_title("This a bar plot project task")
ax.set_xlabel("X axis title")
ax.set_ylabel("Y axis title")
fig
```

```python
for i in ax.patches:
  i.set_edgecolor('black')
  i.set_ls('dashed')
  i.set_lw(3)
  i.set_hatch("\\")
  i.set_facecolor('r')
fig
```

```python
BarBanai(ax, x, y, 0.8)
fig
```

```python
rec_id = range(10)
zorder = reversed(rec_id)

for r, z, in zip(rec_id, zorder):
  ax.patches[r].set_zorder(z)

fig
```

```python
from matplotlib.patches import FancyBboxPatch
```

```python
x = list(range(1,10))
y = x
```

```python
fig, ax = plt.subplots(figsize=(15,5))
def makeFancybar(ax, x, y, w):
  for x, y in zip(x, y):
    fbox = FancyBboxPatch((x,0), w, y, boxstyle = 'round')
    ax.add_patch(fbox)
makeFancybar(ax, x, y, 0.3)
ax.autoscale()
```

# Legends : Add legend

```python
from matplotlib.legend import Legend
```

```python
rec1 = Rectangle((0.5, 0), 0.2, 0.8, label = 'Day1')
rec2 = Rectangle((0.2, 0), 0.2, 0.6, label = 'Day2')
rec3 = Rectangle((0.8, 0), 0.2, 0.4, label = 'Day3')

fig, ax = plt.subplots()
colors = ['red','blue','black']

ax.add_patch(rec1)
ax.add_patch(rec2)
ax.add_patch(rec3)


for i,j in zip (ax.patches, colors):
  i.set_color(j)

handles, labels = ax.get_legend_handles_labels()
legend = Legend(ax, handles = handles, labels = labels)
ax.add_artist(legend)
```

```python
ax.legend()
fig
```

```python
legend.texts[1].set_color('y')
fig
```

```python
legend.legendPatch.set_boxstyle('round, pad = 0.6, rounding_size = 3')
fig
# For more features help(Legend)
```

# Axis Tricks

# Axes Tick Locators and Formatters

```python
fig, ax = plt.subplots()
```

```python
ax.axis([0,6,0,10])
fig
```

```python
ax.set_xlim(0,12)
ax.set_ylim(0,5)
fig
```

```python
x = ax.xaxis
from matplotlib.ticker import AutoLocator, MaxNLocator, LinearLocator, MultipleLocator, FixedLocator, NullLocator
```

```python
auto_loc = AutoLocator()
x.set_major_locator(auto_loc)
fig
```

```python
Maxl = MaxNLocator()
x.set_major_locator(Maxl)
fig
```

```python
lin_l = LinearLocator()
x.set_major_locator(lin_l)
fig
```

```python
Mul_l = MultipleLocator()
x.set_major_locator(Mul_l)
fig
```

```python
Fix_l = FixedLocator(locs = [0,2,4,6,8])
x.set_major_locator(Fix_l)
fig
```

```python
Null_l = NullLocator()
x.set_major_locator(Null_l)
fig
```

```python
x.set_major_locator(auto_loc)
fig
```

```python
from matplotlib.ticker import NullFormatter, FixedFormatter, StrMethodFormatter, ScalarFormatter, PercentFormatter,IndexFormatter
```

```python
Null_f = NullFormatter()
x.set_major_formatter(Null_f)
fig
```

```python
Str_f = StrMethodFormatter("${x:.2f}")
x.set_major_formatter(Str_f)
fig
```

```python
Sca_f = ScalarFormatter()
x.set_major_formatter(Sca_f)
fig
```

```python
Per_f = PercentFormatter(xmax = 12)
x.set_major_formatter(Per_f)
fig
```

```python
Ind_f = IndexFormatter(list("abcdefghijkl"))
x.set_major_formatter(Ind_f)
fig
```
